# Data Cleaning

```python
# Tell which column has null values
df1.isnull().any()
```

```python
# For any column fill the missing value with a mean value
mean_monthly_inc = df1.MonthlyIncome.mean()
mean_dependents =df1.NumberOfDependents.mean()

df1['MonthlyIncome'].fillna(value= mean_monthly_inc, inplace= True)
df1['NumberOfDependents'].fillna(value= mean_dependents, inplace= True)

df1.head(10)
```

```python
# RevolvingUtilizationOfUnsecuredLine Column: No missing value and filter values only b/w 0 and 1

df1['RevolvingUtilizationOfUnsecuredLines'].isnull().any()
```

```python
df1= df1[df1['RevolvingUtilizationOfUnsecuredLines'].between(0,1)]
df1.head()
```

```python
# Cap the age column b/w 18 and 80
df1= df1[df1['age'].between(18,80)]
```

```python
# For column, NumberOfOpenCreditLinesAndLoans take a value less than 30 
df1= df1[df1['NumberOfOpenCreditLinesAndLoans']<30]
```

```python
# For column MonthlyIncome cap it b/w 1000 and 50000
df1= df1[df1['MonthlyIncome'].between(1000,50000)]
df1.head()
```

# Visualization

```python
# visualization for delinquency :
delq = dict(df1.SeriousDlqin2yrs.value_counts())
delq_status=[]
delq_count=[]
for i in delq:
    delq_status.append(i),
    delq_count.append(delq[i])
delq_status[0]='No Serious Delinquency'
delq_status[1]='In serious Delinquency'
plt.figure(figsize=(10,7))
plt.pie(delq_count, labels= delq_status, autopct='%0.1f%%',explode=(0,0.1))
plt.show()
```

```python
# Visualization for RevolvingUtilizationOfUnsecuredLines:-
# note= revolving utilization rate compares your credit debt to your total credit limit.
# Secured loans and lines of credit are secured against your assets, resulting in higher borrowing amount and lower interest rates. 
# Unsecured loans allow for faster approvals since collateral is not required.
# so , for consumer the lower revolving utilization is better. 
# in banking if consumer is already having higher revolving utilization rates , then his repaying capacity is considered bad.

sns.displot(df1['RevolvingUtilizationOfUnsecuredLines'], kde=True, bins=50)
```

```python
# Q3)Create new column NOCLL_Cat from NumberOfOpenCreditLinesAndLoans by :
 #  if value <5 then 1
 #  if 5<= value <8 then 2
 #  if 8<= value <11 then 3
 #  Else then 4
def categorise(row):  
    if row['NumberOfOpenCreditLinesAndLoans'] <5:
        return '1'
    elif row['NumberOfOpenCreditLinesAndLoans'] >=5 and row['NumberOfOpenCreditLinesAndLoans'] <8:
        return '2'
    elif row['NumberOfOpenCreditLinesAndLoans'] >=8  and row['NumberOfOpenCreditLinesAndLoans'] <11:
        return '3'
    return '4'
df1['NOCLL_Cat'] = df1.apply(lambda row: categorise(row), axis=1)
df1.head()
```

```python
# Q4) Visualize NOCLL_Cat as bar graph distribution

sns.displot(df1['NOCLL_Cat'], bins=50, color='violet')
```
